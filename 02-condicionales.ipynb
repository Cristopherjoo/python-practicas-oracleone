10. **Escribe un programa** que lea dos números y luego pregunte a la persona usuaria qué operación desea realizar. El resultado de la operación debe incluir información sobre el número, si es par o impar, positivo o negativo, e entero o decimal.

```python
print('***************')
print('* Calculadora *')
print('***************')

# Función para solicitar un número válido
def pedir_numero():
    while True:
        try:
            dato = float(input("Ingrese un número: "))
            return dato
        except ValueError:
            print("Entrada no válida. Por favor, ingrese un número válido.")

# 1. Ingreso de datos con validación
print("Ingrese el primer número:")
dato_1_valido = pedir_numero()

print("Ingrese el segundo número:")
dato_2_valido = pedir_numero()

# 2. Función para verificar el tipo, paridad y signo
def detalles_resultado(resultado):
    # Detectar si la parte decimal es ".0" si lo es, se identificara como entero
    if resultado.is_integer():
        resultado = int(resultado)  # Convertir a entero si es un número entero

    # Crear un diccionario con los detalles
    resultado_detalles = {
        'resultado': resultado,
        'tipo': 'entero' if isinstance(resultado, int) else 'decimal',
        'paridad': 'par' if resultado % 2 == 0 else 'impar',
        'signo': 'positivo' if resultado > 0 else 'negativo' if resultado < 0 else 'cero'
    }
    return resultado_detalles

# 3. Realizar operaciones
print('Elija la operación que desea realizar:')
print('\tSuma: 1, Resta: 2, Multiplicación: 3, División: 4')

operacion = input('Ingrese el número de la operación a calcular: ')

if operacion == "1":  # Suma
    resultado = dato_1_valido + dato_2_valido
    detalles = detalles_resultado(resultado)
    print(f"Resultado de la suma: {detalles['resultado']} ({detalles['tipo']}, {detalles['paridad']}, {detalles['signo']})")

elif operacion == "2":  # Resta
    resultado = dato_1_valido - dato_2_valido
    detalles = detalles_resultado(resultado)
    print(f"Resultado de la resta: {detalles['resultado']} ({detalles['tipo']}, {detalles['paridad']}, {detalles['signo']})")

elif operacion == "3":  # Multiplicación
    resultado = dato_1_valido * dato_2_valido
    detalles = detalles_resultado(resultado)
    print(f"Resultado de la multiplicación: {detalles['resultado']} ({detalles['tipo']}, {detalles['paridad']}, {detalles['signo']})")

elif operacion == "4":  # División
    if dato_2_valido == 0:
        print("Error: No se puede dividir entre cero.")
    else:
        resultado = dato_1_valido / dato_2_valido
        detalles = detalles_resultado(resultado)
        print(f"Resultado de la división: {detalles['resultado']} ({detalles['tipo']}, {detalles['paridad']}, {detalles['signo']})")

else:
    print("Operación no válida. Por favor elija una opción correcta.")


```

11. **Escribe un programa** que pida a la persona usuaria tres números que representan los lados de un triángulo. El programa debe informar si los valores pueden utilizarse para formar un triángulo y, en caso afirmativo, si es equilátero, isósceles o escaleno. Ten en cuenta algunas sugerencias:
    - Tres lados forman un triángulo cuando la suma de cualesquiera dos lados es mayor que el tercero.
    - **Triángulo Equilátero**: tres lados iguales.
    - **Triángulo Isósceles**: dos lados iguales.
    - **Triángulo Escaleno**: tres lados diferentes.

```python
a = input('Digite primer numero: ')
b = input('Digite segundo numero: ')
c = input('Digite tercer numero: ')

# Verificar si número es entero o decimal
def numeros(valor):
    try:
        # Convertir a float 
        num = float(valor)
        
        # Verificar si es entero (cuando la parte decimal sea .0)
        if num.is_integer():
            return int(num)  # Devolver como entero
        else:
            return num  # Devolver como decimal
    except ValueError:
        return None  # Retornar None si existe un error 

# Verificado y Convertidos
a = numeros(a)
b = numeros(b)
c = numeros(c)

# Verificar que los valores sean números válidos usando none
if a is None or b is None or c is None:
    print("Error: Alguno de los valores ingresados no es un número válido.")
else:
    if a == b == c:  
        print("Triángulo Equilátero.")
    elif a == b or b == c or a == c:  
        print("Triángulo Isósceles.")
    else:  
        print("Triángulo Escaleno.")
```

12. **Escribe un programa** para calcular el valor a pagar por un cliente en una tienda de combustibles. Los descuentos varían según el tipo de combustible y la cantidad comprada.  
    **Reglas**:
    - **Etanol**: hasta 15 litros → 2% de descuento por litro; más de 15 litros → 4% de descuento por litro.  
    - **Diésel**: hasta 15 litros → 3% de descuento por litro; más de 15 litros → 5% de descuento por litro.  
    El precio por litro de diésel es R$ 2,00 y el precio por litro de etanol es R$ 1,70.

```python
combustible = input('¿Qué combustible necesita? "diesel/etanol": ').lower()

# Validación del combustible
if combustible not in ['etanol', 'diesel']:
    print("Opción no válida. Debe elegir entre 'etanol' o 'diesel'.")
else:
    # Descuento Etanol 
    desc_et = 1.70 * 0.98  # 2% 
    desc_15_et = 1.70 * 0.96  # 4% compra más de 15 litros

    # Descuento Diésel
    desc_di = 2.00 * 0.97  # 3% 
    desc_15_di = 2.00 * 0.95  # 5% compra más de 15 litros

    # Variables para almacenar el valor total inicializada en cero 
    descuento_total_etanol = 0
    descuento_total_diesel = 0


    # Opción "etanol"
    if combustible == 'etanol':
        # Ingresar litros de etanol
        litro_etanol = float(input('Ingrese litros de etanol a comprar: '))
        
        # Calcular valor total para Etanol
        if litro_etanol > 0 and litro_etanol <= 15:
            descuento_total_etanol = litro_etanol * desc_et
            print(f"Valor total de Etanol es: R$ {descuento_total_etanol:.2f}")
        elif litro_etanol > 15:
            descuento_total_etanol = litro_etanol * desc_15_et
            print(f"Valor total de Etanol es: R$ {descuento_total_etanol:.2f}")
        else:
            print("Cantidad de Etanol no válida")

    # Opción "diesel"
    elif combustible == 'diesel':
        # Ingresar litros de diésel
        litro_diesel = float(input('Ingrese litros de diesel a comprar: '))
        
        # Calcular valor total para Diésel
        if litro_diesel > 0 and litro_diesel <= 15:
            descuento_total_diesel = litro_diesel * desc_di
            print(f"Valor total de Diésel es: R$ {descuento_total_diesel:.2f}")
        elif litro_diesel > 15:
            descuento_total_diesel = litro_diesel * desc_15_di
            print(f"Valor total de Diésel es: R$ {descuento_total_diesel:.2f}")
        else:
            print("Cantidad de Diésel no válida")

    # Combustible elegido Mostrar el total
    if descuento_total_etanol > 0:
        # Etanol 
        total_pago = descuento_total_etanol
        print(f"El valor total a pagar por Etanol es: R$ {total_pago:.2f}")

    elif descuento_total_diesel > 0:
        # Diésel 
        total_pago = descuento_total_diesel
        print(f"El valor total a pagar por Diésel es: R$ {total_pago:.2f}")

    else:    
        print("Opción no válida o cantidad de combustible incorrecta.")
```
