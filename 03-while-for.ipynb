#### Ejercicio 2 – Crecimiento de colonias bacterianas
Ciclo a usar: while

**Descripción:**  
Escribe un programa para calcular cuántos días tomará que la colonia de una bacteria A supere o iguale a la colonia de una bacteria B, basado en tasas de crecimiento del 3% y 1.5%, respectivamente.  
Supón que la colonia A comienza con 4 elementos y B con 10.




```python
# Diseño algorítmico
'''
Poblaciones iniciales:
    A comienza con 4 bacterias
    B comienza con 10 bacterias

Crecimiento diario:
    A crece un 3% por día
    B crece un 1.5% por día

El bucle:
    Cada día, A se multiplica por 1.03 y B por 1.015
    Contamos los días hasta que la población de A sea igual o mayor que la de B
'''

# Cantidad inicial de bacterias
poblacion_a = 4  # Población inicial A
poblacion_b = 10 # Población inicial B

# Tasas de crecimiento 
TCRECIMIENTO_A = 0.03   # A crece 3% diario
TCRECIMIENTO_B = 0.015  # B crece 1.5% diario

# Factores de crecimiento 
FACTOR_A = 1 + TCRECIMIENTO_A    # 1(100%) + 0.03 = 1.03 --> A se multiplica cada día por 1.03
FACTOR_B = 1 + TCRECIMIENTO_B    # 1(100%) + 0.015 = 1.015 --> B se multiplica cada día por 1.015

dias = 0

# Calcular el crecimiento hasta que A supere a B
while poblacion_a <= poblacion_b:  # Mientras A no sea mayor que B
    poblacion_a *= FACTOR_A  # A se multiplica cada día por el factor de crecimiento
    poblacion_b *= FACTOR_B  # B se multiplica cada día por el factor de crecimiento
    dias += 1 


print(f"Dias necesarios para que la población de A supere a la de B: {dias}")
```

Ejercicio 3 – Validación de calificaciones de usuarios

**Descripción:**  
Para procesar una cantidad de 15 datos de evaluaciones de usuarios de un servicio de la empresa, necesitamos verificar si las calificaciones son válidas.  
Por lo tanto, escribe un programa que recibirá calificaciones del 0 al 5 y verificará si son valores válidos.  
Si se ingresa una calificación superior a 5 o inferior a 0, se repetirá hasta que el usuario ingrese un valor válido.




```python
calificaciones = []
contador = 0

while contador < 15:
    try:
        ingresar_nota = input('Ingrese una calificación entre 0 y 5 (o "exit" para salir)exit')

        # Escribir 'exit', termina el programa
        if ingresar_nota.lower() == 'exit':
            print("Programa terminado.")
            break

        # Convertimos la entrada a flotante
        ingresar_nota = float(ingresar_nota)

        if 0 <= ingresar_nota <= 5:
            contador += 1
            calificaciones.append(ingresar_nota)
        else:
            print('Nota fuera del rango, ingrese una nota válida entre 0 a 5')

    except ValueError:
        # Manejo de error cuando la entrada no es un número
        print('** Error ** Ingresar una calificación numérica entre 0 a 5.')

print(f'Total de notas guardadas: {calificaciones}')
print(f'Total de notas guardadas: {len(calificaciones)}')
```

#### Ejercicio 4 - Promedio de temperaturas

**Descripción:**  
Desarrolla un programa que lea un conjunto indefinido de temperaturas en grados Celsius y calcule su promedio.  
La lectura debe detenerse al ingresar el valor -273°C.



```python
guardar_1 = []
guardar_2 = []
guardar_3 = []
guardar_4 = []
contador = 0

print("Ingrese edades entre 0 y 100. Escriba 'exit' o un número negativo para terminar.")

while True:
    entrada = input("Edad: ")

    if entrada.lower() == "exit":
        print("Programa terminado.")
        break

    try:
        edad = int(entrada)
        
        #Edad negativa
        if edad < 0:
            print("Programa terminado.")
            break

        if 0 <= edad <= 25:
            guardar_1.append(edad)
            print(f"[0-25]: {edad}")
        elif 26 <= edad <= 50:
            guardar_2.append(edad)
            print(f"[26-50]: {edad}")
        elif 51 <= edad <= 75:
            guardar_3.append(edad)
            print(f"[51-75]: {edad}")
        elif 76 <= edad <= 100:
            guardar_4.append(edad)
            print(f"[76-100]: {edad}")
        else:
            print("Edad fuera del rango permitido. Intente de nuevo.")
            continue

        contador += 1

    except ValueError:
        print("Entrada inválida. Debe ser un número o 'exit' para salir.")

# Resultado final
print("\n--- Resumen de edades ingresadas ---")
print(f"[0-25]: {guardar_1}")
print(f"[26-50]: {guardar_2}")
print(f"[51-75]: {guardar_3}")
print(f"[76-100]: {guardar_4}")
print(f"Total de personas registradas: {contador}")
```

#### Ejercicio 5 - Cálculo del factorial

**Descripción:**  
Escribe un programa que calcule el factorial de un número entero proporcionado por el usuario.  
Recuerda que el factorial de un número entero es el producto de ese número por todos sus antecesores hasta llegar al número 1.  
Por ejemplo: el factorial de 5 es 5 x 4 x 3 x 2 x 1 = 120.


```python
def calculo_factorial(n):
    # Inicializamos el resultado como 1
    resultado = 1
    for i in range(1, n + 1):
        resultado *= i  
    return resultado


dato_input = int(input("Ingrese un número para calcular su factorial numero entre 1 a 10"))

if 0<= dato_input <=10:

    # Calcular el factorial
    factorial = calculo_factorial(dato_input)
    
    # Mostrar el resultado
    print(f"El factorial de {dato_input} es {factorial}")
else:
    print('Numero fuera del rango')

```

Proyecto 7 - Número primo

**Descripción:**  
Los números primos tienen diversas aplicaciones en Ciencia de Datos, como en criptografía y seguridad.  
Un número primo es aquel que es divisible solo por sí mismo y por 1.  
Crea un programa que solicite un número entero y determine si es un número primo o no.


```python
num = int(input("Ingrese un número: "))

if num <= 1:
    print(f"{num} no es un número primo.")
else:
    # 2 es el único número primo par
    if num == 2:
        print(f"{num} es un número primo.")  
    elif num % 2 == 0:
        # cualquier otro número par no es primo
        print(f"{num} no es un número primo.")  
    else:
        # Solo probamos divisores impares
        for i in range(3, num):  # Solo probamos divisores impares
            if num % i == 0:
                print(f"{num} no es un número primo.")
                break
        else:
            print(f"{num} es un número primo.")
```

#### Proyecto 9 - Elección de gerencia


**Descripción:**  
En una elección para la gerencia de una empresa con 20 empleados, hay cuatro candidatos.  
Escribe un programa que calcule al ganador de la elección.  

**Detalles:**

- Cada empleado votó por uno de los cuatro candidatos (representados por los números **1, 2, 3 y 4**).
- También se contaron los **votos nulos** (representados por el número **5**) y los **votos en blanco** (representados por el número **6**).

**El programa debe mostrar al final:**

- El número total de votos para cada candidato.
- El número de votos nulos.
- El número de votos en blanco.
- El **porcentaje de votos nulos** respecto al total.
- El **porcentaje de votos en blanco** respecto al total.


```python

import random

empleados = list(range(1, 21))
candidatos = [1, 2, 3, 4, 5, 6]  # 1-4 candidatos, 5=nulo, 6=blanco

votaciones = []

#  voto al azar
for numero in empleados:
    elegido = random.choice(candidatos)
    votaciones.append((numero, elegido))

# Contadores de votos diccionario clave-valor
votos_candidatos = {
    1: 0,  # Candidato 1
    2: 0,  # Candidato 2
    3: 0,  # Candidato 3
    4: 0   # Candidato 4
}
votos_nulos = 0
votos_blancos = 0

# Contamos los votos según su tipo
for _, voto in votaciones:
    if voto in votos_candidatos:
        votos_candidatos[voto] += 1
    elif voto == 5:
        votos_nulos += 1
    elif voto == 6:
        votos_blancos += 1

total_votos = len(votaciones)

# Porcentajes
porcentaje_nulos = (votos_nulos / total_votos) * 100
porcentaje_blancos = (votos_blancos / total_votos) * 100

# Mostrar resultados
print("\n--- Resultados de las votaciones ---")

for candidato, votos in votos_candidatos.items():
    print(f"Candidato {candidato}: {votos} votos")

print(f"Votos nulos: {votos_nulos}")
print(f"Votos en blanco: {votos_blancos}")
print(f"Porcentaje de votos nulos: {porcentaje_nulos:.2f}%")
print(f"Porcentaje de votos en blanco: {porcentaje_blancos:.2f}%")

# Ganador
ganador = max(votos_candidatos, key=votos_candidatos.get)
print(f"El ganador es el Candidato ** {ganador} ** con {votos_candidatos[ganador]} votos.")
```
